```python
import pandas as pd
import joblib
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load your model and label encoder
model = joblib.load('/content/drive/MyDrive/Colab Notebooks/career_model.pkl')
label_encoder = joblib.load('/content/drive/MyDrive/Colab Notebooks/role_encoder.pkl')

# Load test data
df_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/test_dataset.csv')

# Separate features and target
X_test = df_test.drop('Role', axis=1)

# Custom encoding: unseen labels get 0
def safe_encode(label, classes):
    if label in classes:
        return np.where(classes == label)[0][0]
    else:
        print(f"Warning: unseen label in test data: '{label}'. Assigning as 0.")
        return 0

# Apply custom encoding to y_test
y_test = df_test['Role'].apply(lambda x: safe_encode(x, label_encoder.classes_)).values

# Predict
y_pred = model.predict(X_test)

# Evaluation
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))
print("\nðŸ“‹ Classification Report:\n", classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```
